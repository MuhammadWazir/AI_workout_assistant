Experiment with many feature selection criteria to select the best

```python
from sklearn.model_selection import train_test_split

#select the angles as features

X1_train = df_train[['left_side_vector_x', 'left_side_vector_y', 'left_side_vector_z',
       'right_side_vector_x', 'right_side_vector_y', 'right_side_vector_z']]

#select the coordinates as features
X2_train= df_train[[ 'nose_x', 'nose_y', 'nose_z', 'nose_v', 'left_shoulder_x',
       'left_shoulder_y', 'left_shoulder_z', 'left_shoulder_v',
       'right_shoulder_x', 'right_shoulder_y', 'right_shoulder_z',
       'right_shoulder_v', 'right_elbow_x', 'right_elbow_y', 'right_elbow_z',
       'right_elbow_v', 'left_elbow_x', 'left_elbow_y', 'left_elbow_z',
       'left_elbow_v', 'right_wrist_x', 'right_wrist_y', 'right_wrist_z',
       'right_wrist_v', 'left_wrist_x', 'left_wrist_y', 'left_wrist_z',
       'left_wrist_v', 'left_hip_x', 'left_hip_y', 'left_hip_z', 'left_hip_v',
       'right_hip_x', 'right_hip_y', 'right_hip_z', 'right_hip_v']]

#select the coordinates and angles as features
X3_train= df_train[
       ['nose_x', 'nose_y', 'nose_z', 'nose_v', 'left_shoulder_x',
       'left_shoulder_y', 'left_shoulder_z', 'left_shoulder_v',
       'right_shoulder_x', 'right_shoulder_y', 'right_shoulder_z',
       'right_shoulder_v', 'right_elbow_x', 'right_elbow_y', 'right_elbow_z',
       'right_elbow_v', 'left_elbow_x', 'left_elbow_y', 'left_elbow_z',
       'left_elbow_v', 'right_wrist_x', 'right_wrist_y', 'right_wrist_z',
       'right_wrist_v', 'left_wrist_x', 'left_wrist_y', 'left_wrist_z',
       'left_wrist_v', 'left_hip_x', 'left_hip_y', 'left_hip_z', 'left_hip_v',
       'right_hip_x', 'right_hip_y', 'right_hip_z', 'right_hip_v',
       'left_side_vector_x', 'left_side_vector_y', 'left_side_vector_z',
       'right_side_vector_x', 'right_side_vector_y', 'right_side_vector_z']]

Y_train= df_train['label']

X1_test = df_test[['left_side_vector_x', 'left_side_vector_y', 'left_side_vector_z',
       'right_side_vector_x', 'right_side_vector_y', 'right_side_vector_z']]

#select the coordinates as features
X2_test= df_test[[ 'nose_x', 'nose_y', 'nose_z', 'nose_v', 'left_shoulder_x',
       'left_shoulder_y', 'left_shoulder_z', 'left_shoulder_v',
       'right_shoulder_x', 'right_shoulder_y', 'right_shoulder_z',
       'right_shoulder_v', 'right_elbow_x', 'right_elbow_y', 'right_elbow_z',
       'right_elbow_v', 'left_elbow_x', 'left_elbow_y', 'left_elbow_z',
       'left_elbow_v', 'right_wrist_x', 'right_wrist_y', 'right_wrist_z',
       'right_wrist_v', 'left_wrist_x', 'left_wrist_y', 'left_wrist_z',
       'left_wrist_v', 'left_hip_x', 'left_hip_y', 'left_hip_z', 'left_hip_v',
       'right_hip_x', 'right_hip_y', 'right_hip_z', 'right_hip_v']]

#select the coordinates and angles as features
X3_test= df_test[
       ['nose_x', 'nose_y', 'nose_z', 'nose_v', 'left_shoulder_x',
       'left_shoulder_y', 'left_shoulder_z', 'left_shoulder_v',
       'right_shoulder_x', 'right_shoulder_y', 'right_shoulder_z',
       'right_shoulder_v', 'right_elbow_x', 'right_elbow_y', 'right_elbow_z',
       'right_elbow_v', 'left_elbow_x', 'left_elbow_y', 'left_elbow_z',
       'left_elbow_v', 'right_wrist_x', 'right_wrist_y', 'right_wrist_z',
       'right_wrist_v', 'left_wrist_x', 'left_wrist_y', 'left_wrist_z',
       'left_wrist_v', 'left_hip_x', 'left_hip_y', 'left_hip_z', 'left_hip_v',
       'right_hip_x', 'right_hip_y', 'right_hip_z', 'right_hip_v',
       'left_side_vector_x', 'left_side_vector_y', 'left_side_vector_z',
       'right_side_vector_x', 'right_side_vector_y', 'right_side_vector_z']]

Y_test= df_test['label']

```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam

# Encode Y_train and Y_test if they are categorical
label_encoder = LabelEncoder()
Y_train_encoded = label_encoder.fit_transform(Y_train)
Y_test_encoded = label_encoder.transform(Y_test)

def build_keras_dnn(input_dim):
    model = Sequential()
    model.add(Input(shape=(input_dim,)))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(256, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification output

    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

for X_train, X_test, i in zip([X1_train, X2_train, X3_train], [X1_test, X2_test, X3_test], [1, 2, 3]):
    print(f"For the X{i} :\n===========>")

    clf1 = LogisticRegression(solver='lbfgs', random_state=1, max_iter=400)
    clf2 = DecisionTreeClassifier(random_state=1)
    clf3 = GaussianNB()

    # Create and train the Keras neural network for binary classification
    dnn_model = build_keras_dnn(input_dim=X_train.shape[1])
    dnn_model.fit(X_train, Y_train_encoded, epochs=10, batch_size=32, verbose=0)

    # Predict with the Keras DNN model
    dnn_preds = dnn_model.predict(X_test)
    dnn_preds = (dnn_preds > 0.5).astype(int)  # Convert probabilities to binary predictions

    # For Ensemble Voting Classifier (excluding Keras DNN)
    eclf = VotingClassifier(estimators=[
        ('lr', clf1),
        ('dt', clf2),
        ('gnb', clf3)
    ], voting='soft')

    for clf, label in zip([clf1, clf2, clf3, eclf],
                          ['Logistic Regression', 'Decision Tree', 'Gaussian NB', 'Ensemble']):
        clf.fit(X_train, Y_train_encoded)
        Y_pred = clf.predict(X_test)
        accuracy = accuracy_score(Y_test_encoded, Y_pred)
        print(f"Accuracy: {accuracy:.2f} [{label}]")

    # Evaluate the Keras DNN model separately
    dnn_accuracy = accuracy_score(Y_test_encoded, dnn_preds)
    print(f"Accuracy: {dnn_accuracy:.2f} [Keras DNN]")
```
