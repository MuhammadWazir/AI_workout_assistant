Create features representing the sides of the core

```python
df["left_side_vector_x"] = df["left_shoulder_x"] - df["left_hip_x"]
df["left_side_vector_y"] = df["left_shoulder_y"] - df["left_hip_y"]
df["left_side_vector_z"] = df["left_shoulder_z"] - df["left_hip_z"]

df["right_side_vector_x"] = df["right_shoulder_x"] - df["right_hip_x"]
df["right_side_vector_y"] = df["right_shoulder_y"] - df["right_hip_y"]
df["right_side_vector_z"] = df["right_shoulder_z"] - df["right_hip_z"]

df.columns

```

Experiment with many feature selection criteria to select the best

```python
from sklearn.model_selection import train_test_split

#select the angles as features

X1 = df[['left_side_vector_x', 'left_side_vector_y', 'left_side_vector_z',
       'right_side_vector_x', 'right_side_vector_y', 'right_side_vector_z']]

#select the coordinates as features
X2= df[[ 'nose_x', 'nose_y', 'nose_z', 'nose_v', 'left_shoulder_x',
       'left_shoulder_y', 'left_shoulder_z', 'left_shoulder_v',
       'right_shoulder_x', 'right_shoulder_y', 'right_shoulder_z',
       'right_shoulder_v', 'right_elbow_x', 'right_elbow_y', 'right_elbow_z',
       'right_elbow_v', 'left_elbow_x', 'left_elbow_y', 'left_elbow_z',
       'left_elbow_v', 'right_wrist_x', 'right_wrist_y', 'right_wrist_z',
       'right_wrist_v', 'left_wrist_x', 'left_wrist_y', 'left_wrist_z',
       'left_wrist_v', 'left_hip_x', 'left_hip_y', 'left_hip_z', 'left_hip_v',
       'right_hip_x', 'right_hip_y', 'right_hip_z', 'right_hip_v']]

#select the coordinates and angles as features
X3= df[
       ['nose_x', 'nose_y', 'nose_z', 'nose_v', 'left_shoulder_x',
       'left_shoulder_y', 'left_shoulder_z', 'left_shoulder_v',
       'right_shoulder_x', 'right_shoulder_y', 'right_shoulder_z',
       'right_shoulder_v', 'right_elbow_x', 'right_elbow_y', 'right_elbow_z',
       'right_elbow_v', 'left_elbow_x', 'left_elbow_y', 'left_elbow_z',
       'left_elbow_v', 'right_wrist_x', 'right_wrist_y', 'right_wrist_z',
       'right_wrist_v', 'left_wrist_x', 'left_wrist_y', 'left_wrist_z',
       'left_wrist_v', 'left_hip_x', 'left_hip_y', 'left_hip_z', 'left_hip_v',
       'right_hip_x', 'right_hip_y', 'right_hip_z', 'right_hip_v',
       'left_side_vector_x', 'left_side_vector_y', 'left_side_vector_z',
       'right_side_vector_x', 'right_side_vector_y', 'right_side_vector_z']]

Y= df['label']

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

for X, i in zip([X1, X2, X3], [1, 2, 3]):
    print("For the X", i, ":\n===========>")

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=43)

    clf1 = LogisticRegression(solver='lbfgs', random_state=1, max_iter=400)
    clf2 = DecisionTreeClassifier(random_state=1)
    clf3 = GaussianNB()
    dnn1 = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=500, random_state=1)
    dnn2 = MLPClassifier(hidden_layer_sizes=(50, 50, 50), max_iter=500, random_state=1)
    eclf = VotingClassifier(estimators=[
        ('lr', clf1),
        ('dt', clf2),
        ('gnb', clf3),
        ('dnn1', dnn1),
        ('dnn2', dnn2)
    ], voting='soft')

    for clf, label in zip([clf1, clf2, clf3, dnn1, dnn2, eclf],
                          ['Logistic Regression', 'Decision Tree', 'Gaussian NB', 'DNN1', 'DNN2', 'Ensemble']):
        clf.fit(X_train, Y_train)
        Y_pred = clf.predict(X_test)
        accuracy = accuracy_score(Y_test, Y_pred)
        print(f"Accuracy: {accuracy:.2f} [{label}]")


```

```python
import pickle

# Fit the DNN1 model on X1
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X1, Y, test_size=0.2, random_state=43)
dnn1 = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=500, random_state=1)
dnn1.fit(X_train1, Y_train1)

# Save the model to a file using pickle
with open('dnn1_X1_model.pkl', 'wb') as file:
    pickle.dump(dnn1, file)
```
